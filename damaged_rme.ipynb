```python
header_cells = repair_sheet.range('CL1:EK1').value  
new_header = np.array(header_cells)
```

```python

def local_repairs(ws, repairs_country_cells, new_header):
    country_weekly_sum = {}

    for country in list(repairs_country_cells.keys()):
        country_start = repairs_country_cells[country][0]
        country_end = repairs_country_cells[country][1]
        v1 = ws.range('CL' + country_start +':'+'EK' + country_end).value                         # range for current country value
        data = pd.DataFrame(v1)                                                                   # converting to dataframe for easy transformation

        data.columns = new_header   
        data = data.fillna(0)
        weekly_sum = data.sum(axis=0)
        country_weekly_sum[country] = weekly_sum

    return country_weekly_sum
```

```python
def local_scraped(ws, scrapped_country_cells, new_header):
    country_weekly_sum = {}
    for country in list(scrapped_country_cells.keys()):
        country_start = scrapped_country_cells[country][0]
        country_end = scrapped_country_cells[country][1]
        v1 = ws.range('CK' + country_start +':'+'EJ' + country_end).value                         # range for current country value
        data = pd.DataFrame(v1)                                                                   # converting to dataframe for easy transformation

        data.columns = new_header   
        data = data.fillna(0)
        weekly_sum = data.sum(axis=0)
        country_weekly_sum[country] = weekly_sum

    return country_weekly_sum
```
